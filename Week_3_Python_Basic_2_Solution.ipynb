# 2.리스트 (주문 처리하기)

## 2.1 리스트

지난 주 만든 키오스크는 주문을 받고 결제를 하고 프로그램이 종료.

실제로는 결제가 이루어진 후 주문된 내용대로 커피를 만들어야 함.

그런데, 지난 주에 만든 키오스크는 한번에 하나의 주문 밖에 받을 수가 없음. 이유는 우리가 사용한 menu 라는 변수에는 하나의 값만 담을 수 있기 때문.

그래서 여러 개의 값을 저장하기 위해 리스트(List)를 사용해보고자 함.

리스트(list)는 여러개의 변수를 저장할 수 있는 자료형.

변수와 마찬가지로 = 을 사용해서 저장하는데, 한가지 다른 점은 오른쪽에 `'[]'` 대괄호를 써 주어야 한다.

예를 들어, `menu = ['아메리카노', '카페라떼', '카페모카', '핫초코', '페퍼민트', '카모마일']`처럼 저장한다. 

저장된 변수를 출력해보자.

````python
menu = ['아메리카노', '카페라떼', '카페모카', '핫초코', '페퍼민트', '카모마일']
print(menu)
````

## 2.2 리스트 인덱싱

만약에 menu 중에서 페퍼민트만 출력하고 싶으면 어떻게 해야할까? 

즉, 리스트 안의 특정 요소(element)에 접근하는 인덱스를 사용해서 가능하다. 리스트 안의 모든 요소는 인덱스를 가지고 있는데, 첫번째 것이 0, 두번째가 1, 이런 순서로 인덱스가 증가한다.

따라서 맨 처음으로 저장된 아메리카노는  `menu[0]`으로 가져올 수 있다.

컴퓨터 프로그래밍에서 인덱스는 대부분 항상 0부터 시작 (R 제외) 

````python
print(menu[0])
````

페퍼민트의 인덱스는? 다섯번째니까 5가 아니라 하나 뺀 4. (n-1)

````python
print(menu[4])
````

위의 리스트에서 가장 마지막 요소는 카모마일로 인덱스는 5.

그런데 만약 6번째 인덱스를 불러오고자 하면 어떻게 될까?

예) `menu[6]` 

````python
print(menu[6])
````

에러가 발생.

### IndexError: list index out of range

리스트의 인덱스가 범위를 벗어났다고 함. 리스트가 가진 범위 0~5까지인데 6을 불러내서 생기는 에러.

이렇게 리스트에서 내가 원하는 요소만 접근(인덱싱)해서 출력할 수 있다. 

파이썬은 음수 인덱싱도 제공한다. `menu[-1]`, `menu[-2]` 와 같은 표현들이 종종 사용된다. 

`menu[-1]` 하면 어떤 값이 출력될까?

````python
print(menu)
print(menu[-1])
print(menu[-2])
print(menu[-3])
````

맨 마지막 요소가 출력된다. `menu[-2]`는? 뒤에서 두번째, 즉 페퍼민트를 불러온다.

이번에는 어떤 범위에 있는 값만 출력해 보자. 예를 들면 아메리카노, 카페라떼, 카페모카 까지만 출력을 해보자. 이런 걸 range 라고 하는데 우리가 일상에서 range 를 표현할 때는 보통 0~2 와 같은 형식의 표기법을 사용한다.

그렇다면 `menu[0~2]` 는 어떤 결과가 출력될까?

````python
print(menu[0~2])
````

유효한 문법이 아니라 에러가 발생. 진짜 맞는 문법은 `menu[0:2]`. 0부터 2까지를 `:` 을 사이에 두고 표시한다. 

````python
print(menu[0:2])
````

예상과는 달리 카페라떼까지만 출력을 함.

0부터 2까지 인데 왜 2번째 인덱스에 저장되어 있는 3번째 요소는 못 가져올까?

그 이유는 파이썬 문법에 range를 `[시작하는 인덱스:끝나는 인덱스 이전까지]` 로 정의되어 있기 때문. 

````python
print(menu[0:3])
````

리스트 일부를 떼어내서 새로운 리스트로 저장하는 것도 가능하다. 커피류와 비커피류로 나눠보자.

````python
coffee = menu[0:3]
noncoffee = menu[3:6]
print(coffee)
print(noncoffee)
````

문자열만 되는게 아니라 당연히 숫자도 가능.

````python
odd = [1,3,5,7,9]
even = [2,4,6,8,10]
mixed = [1, 2, 3, "one", "two", "1", "2"]

print(odd)
print(even)
print(mixed)
````

## 2.4 리스트 정렬

위에서 리스트를 합쳤을 때 안의 요소들이 순서대로 정렬되지 않았다. 만약 필요에 의해 리스트 내용을 정렬하고자 한다면 다음의 두가지 방법을 사용할 수 있다.
```
1. list 뒤에 .sort()를 해서 정렬하는 방법 
2. sorted() 괄호 안에 리스트를 넣는 방법
```

`sort()`는 원래 리스트의 내용 자체를 바꿔주지만 `sorted()`는 원래 리스트의 내용은 바꾸지 않음.

````python
uptoten = odd+even
uptoten.sort()
print(uptoten)
````

````python
uptoten = odd+even
print(sorted(uptoten))
print(uptoten)
````

## 5.1 While loop

앞의 단계에서는 주문을 할 때 하나의 주문 밖에 하지 못했다.


여기서 사용되는 조건은 종료조건이다. 

while 문은 종료할 수 있는 조건이 없다면 **무한반복** 되니 사용에 주의해야 한다.

````python
while (종료조건):
  실행문
````

## 5.2 참과 거짓 (True and False)

while 문을 이용하면 무한히 프로그램이 돌아가도록 만들 수 있다.

사실 우리가 쓰는 대부분의 프로그램은 특정한 조건이 충족되기 전에는 무한히 돌아가는 프로그램이라고 볼 수 있다.

while 문은 조건이 만족되면 (True) 계속 반복하며 while 문 블락 아래에 있는 코드를 실행한다.

즉, 조건이 True 라면, while 문은 이른바 무한루프에 빠진다.
다음의 코드를 실행해보자.

````python
while True:
  print('무한 실행')
  
````

따라서 무한루프에 빠진 프로그램을 종료시키기 위해서는 종료 조건이 필요하다. 보통 다음과 같이 작성한다.


````python
x = 0
while x < 10:
    print(x, "is less than 10")
    x += 1 
````

While 문을 이용하면 여러 건의 주문을 받을 수 있다.

코드를 고치기 전 예전 코드를 다시 살펴보자.

````python
# 메뉴판
print('안녕하세요 스누카페입니다.')
print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
print('---------- 메뉴 ----------')
print('아메리카노 1000원')
print('카페라떼  1500원')

# 주문
menu = input('무엇을 주문하시겠습니까?\n')
amount = input('몇 잔 드릴까요?\n')

# 주문확인
print(menu + ' ' + amount + '잔 준비하겠습니다.')

# 가격 계산
if menu == '아메리카노':
  price = 1000 * int(amount)
else:
  price = 1500 * int(amount)
print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

# 결제
payment = input('결제는 어떻게 도와드릴까요?\n')
if payment == '카드':
  print('카드리더기에 카드를 넣어주세요.')
else:
  print('투입구에 현금을 넣어주세요 ')
````

위의 코드 중에서 여러 번 반복적으로 진행되어야 할 곳은 어디일까? 주문을 받는 곳이다. 

그럼 while 을 이용하여 주문 받는 곳을 고쳐보자.

````python
#@title
print('안녕하세요 스누카페입니다.')
print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
print('---------- 메뉴 ----------')
print('아메리카노 1000원')
print('카페라떼  1500원')

## 수정된 부분 ------------------------------------
while(True):
  menu = input('무엇을 주문하시겠습니까?\n')
  amount = input('몇 잔 드릴까요?\n')
## ---------------------------------------------

print(menu + ' ' + amount + '잔 준비하겠습니다.')
if menu == '아메리카노':
  price = 1000 * int(amount)
else:
  price = 1500 * int(amount)
print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

payment = input('결제는 어떻게 도와드릴까요?\n')
if payment == '카드':
  print('카드리더기에 카드를 넣어주세요.')
else:
  print('투입구에 현금을 넣어주세요 ')
````

주문은 계속 받을 수 있지만, 그 아래 코드로 진행되지 못하고 무한루프에 빠진다. 왜일까? 

## 5.3 While문 조건


````python
while (조건):
  실행문
````

while 문의 종료조건을 만드는 방법은 다양하다.

여기서는 준비된 커피가 다 떨어질 때 까지 주문을 받는 경우를 생각해 보자.

커피의 잔수를 10잔부터 0잔까지 카운트다운하는 코드를 만들어 보자.

그러기 위해서는 우선 시작조건을 만들어야 한다.

`coffee = 10`

시작조건은 변수의 선언과 할당에서 시작한다.

이제 이 시작조건이 loop가 진행되면서 변화할 수 있도록 코드를 짜줘야 한다.

한잔 씩 감소하도록 코드를 짜보자. 어떻게 할 수 있을까?

`coffee = coffee - 1`

현재의 커피 잔수에서 1을 빼서 coffee 라는 변수에 재할당한다. 

그렇다면 10이었던 coffee 의 값이 9로 감소할 것이다.

이 코드가 while 문 안에 들어있다면 coffee 의 값은 매번 반복할 때마다 감소할 것이다.

그런데, **종료조건**이 없다면 이 반복문은 무한루프에 빠진다.

종료가 되기위한 조건은 무엇일까?

아마도 커피가 다 떨어지면 더 팔수가 없으니 이것을 종료조건으로 두어야할 것 같다. 

즉, 코드로 설명하면 `coffee == 0` 이 종료조건이 되는 것이다. 

그렇다면 `coffee == 0` 이 종료조건인가? 

while 문은 조건을 만족하지 않을 때까지 반복이 되기 때문에 종료조건은 다음처럼 작성하도록 하자.

`coffee > 0`

코드로 구현해보자.

````python
coffee = 10

while(coffee > 0):
  print("커피가 {}잔 남았습니다.".format(coffee))
  coffee = coffee - 1
````

다시 우리의 커피숍 예제로 돌아와 예전 코드를 보자.

주문 받는 코드는 다음과 같다.

````python
while(True):
  menu = input('무엇을 주문하시겠습니까?\n')
  amount = input('몇 잔 드릴까요?\n')
````

## 5.4 Break

그래도 아직은 어설프다. 계속 "몇 잔 드릴까요?" 라고 물어보는게 어색하다. 

"추가 주문 하시겠어요?"라고 물어보는 편이 훨씬 나을 것 같다. 

그리고 만약 고객이 싫다고 하면 while 루프에서 벗어나야 할 것이다.

이때 사용하는 명령어가 break 다. break는 반복문을 빠져나올 때 사용한다.

즉, while문에서 특정 조건이 발생하면 while문의 조건과 상관없이 break로 빠져나올 수 있다.

````python
americano = 0

while(True):
  if americano > 0:
    print('아메리카노 주문이 가능합니다.')
  else:
    print('아메리카노가 품절되었습니다.')
    break
````

break를 이용해서 코드를 고쳐보자. if문을 써서 추가주문을 할 것인지 묻고, 아니라고 말하면 break로 빠져나오는 것입니다. 추가 주문의 변수명은 additionalOrder라고 합시다.

````python
#@title
# 메뉴판
print('안녕하세요 스누카페입니다.')
print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
print('---------- 메뉴 ----------')
print('아메리카노 1000원')

# 아메리카노 수량
americano = 2

# 주문 받기
while(americano > 0):
  menu = '아메리카노'
  amount = input('몇 잔 드릴까요?\n')
  americano = americano - int(amount)
  additionalOrder = input('추가 주문하시겠어요?\n')
  if additionalOrder == '아니요':
    break

# 주문 확인
print(menu + ' ' + amount + '잔 준비하겠습니다.')

# 가격 게산
price = 1000 * int(amount)
print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

# 결제
payment = input('결제는 어떻게 도와드릴까요?\n')
if payment == '카드':
  print('카드리더기에 카드를 넣어주세요.')
else:
  print('투입구에 현금을 넣어주세요 ')
````

아까보다는 나아졌지만 2잔(max)를 주문하면 끝나야 하는데 계속 추가 주문을 물어보는게 어색하다. 

새로운 조건문을 이용하여 이 부분을 해결해보자. americano 가 0잔이 되면 품절이라고 말하는 코드가 들어가면 좋을 것 같다.

````python
#@title
# 메뉴판
print('안녕하세요 스누카페입니다.')
print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
print('---------- 메뉴 ----------')
print('아메리카노 1000원')

# 아메리카노 수량
americano = 2

# 주문 받기
while(americano > 0):
  menu = '아메리카노'
  amount = input('몇 잔 드릴까요?\n')
  americano = americano - int(amount)
  if americano < 1:
    print('아메리카노가 품절되었습니다.')
    break
  additionalOrder = input('추가 주문하시겠어요?\n')
  if additionalOrder in ['아니요', '아니오', 'no', 'NO', 'X']:
    break

# 주문 확인
print(menu + ' ' + amount + '잔 준비하겠습니다.')

# 가격 계산
price = 1000 * int(amount)
print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

# 결제
payment = input('결제는 어떻게 도와드릴까요?\n')
if payment == '카드':
  print('카드리더기에 카드를 넣어주세요.')
else:
  print('투입구에 현금을 넣어주세요 ')
````

지금까지 만든 코드는 주문을 받을 때만 반복을 하고 주문이 끝나 결제까지 이루어지면 프로그램이 종료된다.

그러나 실제 프로그램은 다음 주문을 위해서라도 계속 돌아가고 있어야 한다. 어떻게 프로그램이 종료되지 않고 계속 주문을 받을 수 있을까?

전체 코드를 while문을 사용해 계속 반복하게 하면 어떨까?

````python
#@title 
while(True):
  # 메뉴판
  print('안녕하세요 스누카페입니다.')
  print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
  print('---------- 메뉴 ----------')
  print('아메리카노 1000원')

  # 아메리카노 수량
  americano = 2

  # 주문 받기
  while(americano > 0):
    menu = '아메리카노'
    amount = input('몇 잔 드릴까요?\n')
    americano = americano - int(amount)
    if americano < 1:
      print('아메리카노가 품절되었습니다.')
      break
    additionalOrder = input('추가 주문하시겠어요?\n')
    if additionalOrder == '아니요':
      break

  # 주문 확인
  print(menu + ' ' + amount + '잔 준비하겠습니다.')

  # 가격 계산
  price = 1000 * int(amount)
  print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

  # 결제
  payment = input('결제는 어떻게 도와드릴까요?\n')
  if payment == '카드':
    print('카드리더기에 카드를 넣어주세요.')
  else:
    print('투입구에 현금을 넣어주세요 ')
````

뭔가 이상하다. 품절된 커피를 계속 주고 있다. 커피가 품절되면 프로그램이 자동으로 멈추도록 해야한다. break 를 사용해보자.

그리고 또 하나 중요한 것. americano 변수가 while 문 안에 있으면 반복문이 한번 돌때마다 americano 의 값이 2로 리셋된다. 그렇기 때문에 무한 주문을 받고 있는 것이다. 

따라서 while 문 밖으로 americano 변수를 빼줄 필요가 있다.

````python
#@title
americano = 2
while(True):
  # 메뉴판
  print('안녕하세요 스누카페입니다.')
  print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
  print('---------- 메뉴 ----------')
  print('아메리카노 1000원')

  # 품절 조건
  if americano < 1:
      print('아메리카노가 품절되었습니다.')
      break
  
  # 주문 받기
  while(americano > 0):
    menu = '아메리카노'
    amount = input('몇 잔 드릴까요?\n')
    americano = americano - int(amount)
    additionalOrder = input('추가 주문하시겠어요?\n')
    if additionalOrder == '아니요':
      break

  # 주문 확인
  print(menu + ' ' + amount + '잔 준비하겠습니다.')

  # 가격 계산
  price = 1000 * int(amount)
  print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

  # 결제
  payment = input('결제는 어떻게 도와드릴까요?\n')
  if payment == '카드':
    print('카드리더기에 카드를 넣어주세요.')
  else:
    print('투입구에 현금을 넣어주세요 ')
````

이번에는 주문을 확인받는 부분을 고쳐보자. 지금의 코드는 주문을 여러 번 받아도 가장 최근 주문만 확인하고 계산하고 있다. 앞에서 배운 튜플을 이용해 메뉴와 수량을 주문 리스트에 넣어주자.

이를 위해서 리스트의 멤버함수(method)인 `.append()`를 사용하려고 한다.

## 5.6 결과물 4 (여러 메뉴 주문 여러 번 받기)

````python
# 음료 수량
americano = 2
caffelatter = 2
while(True):
  # 메뉴판
  print('안녕하세요 스누카페입니다.')

  # 품절 안내
  if americano < 1 and caffelatter < 1:
    print('죄송하지만, 모든 메뉴가 품절되었습니다.')
    break

  # 메뉴판
  print('저희 카페에서 준비한 메뉴는 다음과 같습니다.')
  print('---------- 메뉴 ----------')
  print('아메리카노 1000원')
  print('카페라떼  1500원')

  menu = ''
  amount = 0
  orders = []
  price = 0

  
  # 주문 받기
  while(americano > 0 or caffelatter > 0):
    if americano < 1:
      print('죄송하지만, 아메리카노가 품절되었습니다.')
    elif caffelatter < 1:
      print('죄송하지만, 카페라떼가 품절되었습니다.')
    menu = input('무엇을 주문하시겠습니까?\n')
    amount = input('몇 잔 드릴까요?\n')
    if menu == '아메리카노' and int(amount) > americano:
      print('현재 아메리카노는 ' + str(americano) +'잔만 주문 가능합니다.')
      continue
    elif menu == '카페라떼' and int(amount) > caffelatter:
      print('현재 카페라떼는 ' + str(caffelatter) +'잔만 주문 가능합니다.')
      continue
    additional = input('추가 주문하시겠습니까?\n')
    if menu == '아메리카노' and americano > 0:
      americano -= int(amount)
    elif menu == '카페라떼' and caffelatter > 0:
      caffelatter -= int(amount)
    orders.append((menu, amount))
    if americano < 1 and caffelatter < 1:
      print('죄송하지만, 모든 메뉴가 품절되었습니다.')
      break
    if additional == '아니요':
      break

  
  # 주문 확인
  for order in orders:
    print(order[0] + ' ' + str(order[1]) + '잔 준비하겠습니다.')

  # 가격 계산
  for order in orders:
    if order[0] == '아메리카노':
      price += 1000 * int(order[1])
    elif order[0] == '카페라떼':
      price += 1500 * int(order[1])
  print('주문하신 메뉴의 가격은 ' + str(price) + '원 입니다.')

  # 결제
  payment = input('결제는 어떻게 도와드릴까요?\n')
  if payment == '카드':
    print('카드리더기에 카드를 넣어주세요.')
  else:
    print('투입구에 현금을 넣어주세요 ')

  # 음료 전달
  for order in orders:
    print('손님, ' + order[0] + ' ' + str(order[1]) + '잔 나왔습니다.')
  
  print('감사합니다.')
````

````python
s = set()
s.add(1)
s.add(2)
s.add(2)
print(s)
````

````python
# set 선언
my_set = set([1,2,3])
print(my_set)

# 값 하나 추가 : add(x)
my_set.add(4)
print(my_set)
 
# 값 여러개 추가 : update(x)
my_set.update([5,6,7,8,9,10])
print(my_set)
 
# 값 제거 : remove(x)
my_set.remove(10)
print(my_set)

# 값 제거 : remove(x)
my_set.remove(6)
print(my_set)
````

set은 순서가 없어서 index를 이용하여 자료에 접근할 수 없다. 따라서 set에서 특정 자료에 접근하기 위해선 list나 tuple로 변환하여야 한다.

````python
# set 선언
my_set = set([1,2,3])
my_list = list(my_set)
my_tuple = tuple(my_set)
 
print(my_list)
print(my_list[1])
print(my_tuple)
print(my_tuple[2])
````

````python
# set 선언
set1 = set([1,2,3])
set2 = set([2,4,5,6])
set3 = set() # 공집합
 
# set1과 set2의 교집합
print(set1 & set2)
print(set1.intersection(set2))
 
# set1과 set2의 합집합
print(set1 | set2)
print(set1.union(set2))
 
# set1과 set2의 차집합
print(set1 - set2)
print(set1.difference(set2))
print(set2.difference(set1))
 
# set1과 set2의 합집합에서 교집합을 뺀 차집합
print(set1 ^ set2)
````

## 7.4 Exceptions

프로그래밍을 하다보면 데이터 연산과 관련하여 항상 에러가 발생하게 된다. 에러의 유형은 여러가지인데, 에러가 발생하면 프로그램은 일반적으로 종료된다. 에러가 발생해도 프로그램이 종료되지 않고 에러를 처리한 후 계속 진행하기 위해서는 exception 이라는 기능을 사용한다. 예컨대 0/0을 수행하면 ZeroDivisionError라는 것이 뜨면서 프로그램이 종료되는데, exception 처리를 해주면 에러를 처리한 후 프로그램은 계속 수행된다. 다음의 코드를 확인해보자.

````python
print(0/0)
print("계산이 완료되었습니다. 다음 계산을 수행해 주세요.")
````

````python
try: 
    print(0/0)
except ZeroDivisionError:
    print("cannot divide by zero")

print("계산이 완료되었습니다. 다음 계산을 수행해 주세요.")
````

데이터 분석을 할 때 random number를 사용하게 되는 경우가 많다. 이때는 random 모듈을 사용한다.

````python
import random
random.random()  # 0-1 사이의 난수를 생성한다.
````

````python
# four_uniform_random = [random.random() for x in range(4)]  # 4개의 난수를 생성하여 리스트에 저장
# print(four_uniform_random)

four_uniform_random = []
for x in range(4):
  four_uniform_random.append(random.random())
    
print(four_uniform_random)
````

for문 등을 사용할 때 의미없이 사용되는 변수는 ``` _ ``` 로 표현한다. 예를 들어 위의 코드에서 ``` for x in range(4) ```에서 x는 외부에서 참조하거나 계산 등에 사용되지 않는다. 이런 경우 파이썬에선 일반적을 ``` _ ``` 로 대체한다.

````python
four_uniform_random = [random.random() for _ in range(4)]  # 4개의 난수를 생성하여 리스트에 저장
print(four_uniform_random)
````

### `readlines()` 함수의 사용

`readlines()`는 문단 별로 (\n 단위로) 문장을 라인으로 읽어서 리스트로 저장한다. 따라서 `for ... in` 등을 사용하여 리스트에 저장된 문장을 꺼낼 수 있다.

````python
# 문단 사이의 공백을 제거하기 위해서는 strip() 함수를 사용한다.
for line in text:
  print(line.strip())
````

## 8.2 File Write

파이썬에서 파일을 저장할때도 `open()`함수를 사용한다. 다만, 사용하는 옵션은 `"w"` 혹은 `"a"`다.

````python
mylist = ["hello", "python", 1, 2, 3]
f = open("/content/drive/MyDrive/_SocialComp_2022-Solution/Week3/result-w.txt", "w")  # w 옵션을 파일을 저장할 때 사용한다.
for i in mylist:
  f.write(str(i) + "\n")
f.close()  # close()를 하지 않으면 파일의 내용이 저장되지 않는다.
````

`"w"` 옵션은 파일의 내용을 대체. 따라서 새로운 내용이 write 되면 이전의 내용은 사라진다.

````python
f = open("/content/drive/MyDrive/_SocialComp_2022-Solution/Week3/result-w.txt", "w")
f.write("이렇게")
f.write("\n")
f.write("값이 ")
f.write("바뀜")
f.close()
````

기존의 내용에 추가로 내용을 작성하기 위해서는 `"a"` 옵션을 사용한다. (append)

````python
f = open("/content/drive/MyDrive/_SocialComp_2022-Solution/Week3/result-w.txt","a")
f.write("\n실행하면 값이 추가됨")
f.close()
````

File Read 와 마찬가지로 ```with open...```을 사용할 수 있습니다. `close()`를 하지 않아도 블록이 끝난 후 자동으로 파일이 `close()`된다.

````python
mylist = ["hello", "python", 1, 2, 3]
with open("/content/drive/MyDrive/_SocialComp_2022-Solution/Week3/result-w.txt", "a") as f:
  for i in mylist:
    f.write(str(i) + "\n")
````
